```python
import pandas as pd
from sklearn.cross_decomposition import CCA
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('complete_filled_data.csv')

# Separate the two groups of variables
pollution_vars = ['Access to clean fuels and technologies for cooking (% of population) [EG.CFT.ACCS.ZS]', 'CO2 emissions (metric tons per capita) [EN.ATM.CO2E.PC]', 'CO2 emissions (kg per PPP $ of GDP) [EN.ATM.CO2E.PP.GD]', 'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]', 'Access to electricity (% of population) [EG.ELC.ACCS.ZS]']
economics_vars = ['GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]', 'GNI (current US$) [NY.GNP.MKTP.CD]', 'Exports of goods and services (% of GDP) [NE.EXP.GNFS.ZS]', 'Imports of goods and services (current US$) [NE.IMP.GNFS.CD]', 'Exports of goods and services (current US$) [NE.EXP.GNFS.CD]']

# Separate the Income Group variable
income_group = df['Income Group']

# Fit the CCA model
cca = CCA(n_components=2)
cca.fit(df[pollution_vars], df[economics_vars])

# Get the canonical correlations and canonical weights
#ccorrs = cca.correlations_
cweights_x = cca.x_weights_
cweights_y = cca.y_weights_

# Project the data onto the canonical space
x_c, y_c = cca.transform(df[pollution_vars], df[economics_vars])

# Visualize the results
plt.scatter(x_c[:, 0], x_c[:, 1], c=income_group.map({'L': 'red', 'UM': 'blue', 'H': 'green', 'LM': 'orange'}))
plt.scatter(y_c[:, 0], y_c[:, 1], c=income_group.map({'L': 'red', 'UM': 'blue', 'H': 'green', 'LM': 'orange'}))
plt.xlabel('Pollution')
plt.ylabel('Economics')
plt.show()
#a scatter plot of the data projected onto the two canonical components, with the points colored according to the Income Group variable.





```

```python
import pandas as pd
import boto
```

```python
from sklearn.cross_decomposition import CCA
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X = data[retained_var_pol]
```

```python
Y = data[retained_var_ec]
```

```python
X_scaled = scaler.fit_transform(X)

Y_scaled = scaler.fit_transform(Y)
```

```python
# Instantiate the Canonical Correlation Analysis with 2 components
my_cca = CCA(n_components=2)
```

```python
# Fit the model
my_cca.fit(X_scaled, Y_scaled)
```

```python
# get canonical correlation coefficient
print(my_cca.coef_)
```

```python
# transform X and Y using the fitted CCA model
X_c, Y_c = my_cca.transform(X, Y)
```

```python
# print transformed data
print(X_c)
print(Y_c)
```

```python
data[['cca_pol_x', 'cca_pol_y']] = X_c
```

```python
data[['cca_ec_x', 'cca_ec_y']] = Y_c
```

```python
data
```

```python
# plot the transformed data with colors
plt.scatter(X_c[:, 0], X_c[:, 1], c=data['Income Group'], cmap='coolwarm')
plt.scatter(Y_c[:, 0], Y_c[:, 1], c=data['Income Group'], cmap='coolwarm')

# add a colorbar to show the range of the third variable
cbar = plt.colorbar()
cbar.ax.set_ylabel('Z')

# add axis labels and a title to the plot
plt.xlabel('CCA Component 1')
plt.ylabel('CCA Component 2')
plt.title('Transformed Data with Colored Points')

# show the plot
plt.show()
```

```python
# Obtain the rotation matrices
xrot = my_cca.x_rotations_
yrot = my_cca.y_rotations_
```

```python
import matplotlib.pyplot as plt

# create an array of ordinal categorical labels for each observation
labels = data['Income Group']

# define a dictionary of colors for each category
colors = {'H': 'r', 'UM': 'g', 'LM': 'b', 'L': 'y'}

# plot the transformed data with colors
for label in set(labels):
    idx = (labels == label)
    plt.scatter(X_c[idx, 0], X_c[idx, 1], c=colors[label], label=label)
    plt.scatter(Y_c[idx, 0], Y_c[idx, 1], c=colors[label], label=label)

# add a legend and colorbar to the plot
plt.legend()
cbar = plt.colorbar()
cbar.ax.set_ylabel('r')

# add axis labels and a title to the plot
plt.xlabel('CCA Component 1')
plt.ylabel('CCA Component 2')
plt.title('Transformed Data with Colored Points')

# show the plot
plt.show()
```
