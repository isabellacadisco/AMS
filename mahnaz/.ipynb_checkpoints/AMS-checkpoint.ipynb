```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
ds = pd.read_csv("D:/wdi2018.csv")
```

```python
#delete the variables 
v=[]
v = pd.DataFrame([v])
k=ds.isna().sum()
i = 0
for i in range(len(ds.columns)):
    if ((k[i] / len(ds.iloc[:,i])) * 100) > 20:
          v.loc[0,i]=i
for i in reversed(range(len(v.columns))):
    ds=ds.drop(ds.columns[v.iloc[0,i]], axis=1)
```

```python
#joining two tables
ds2 = pd.read_csv('D:/income.csv')
ds.loc[:,448]=0
for i in range(len(ds)):
    for j in range(len(ds2)):
        if ds2.iloc[j,0] == ds.iloc[i,3]:
            ds.loc[i,448]= ds2.iloc[j,2]
```

```python
#outlier for the pollution variables
pollution_data = dataset[retained_var_pol]
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from collections import Counter
from sklearn.neighbors import NearestNeighbors
```

```python
df = StandardScaler().fit_transform(pollution_data)

neighbors = NearestNeighbors(n_neighbors=10, metric= 'mahalanobis')
neighbors_fit = neighbors.fit(df)
distances, indices = neighbors_fit.kneighbors(df)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
fit = DBSCAN(eps=5, min_samples=10).fit(df)
outliers_ = pd.DataFrame(df) 
outliers = outliers_[fit.labels_== -1]
print(Counter(fit.labels_))
```

```python
outliers
```

```python
#outliers for economic variables
econ_data = dataset[retained_var_ec]
```

```python
df2 = StandardScaler().fit_transform(econ_data)
neighbors = NearestNeighbors(n_neighbors=10, metric= 'mahalanobis')
neighbors_fit = neighbors.fit(df2)
distances, indices = neighbors_fit.kneighbors(df2)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
fit = DBSCAN(eps=7, min_samples=10).fit(df2)
outliers_ = pd.DataFrame(df2) 
outliers = outliers_[fit.labels_== -1]
print(Counter(fit.labels_))
```

```python
import math
from scipy.stats import shapiro 
from pingouin import multivariate_normality
from statsmodels.multivariate.manova import MANOVA
```

```python
multivariate_normality(dataset[retained_var_pol], alpha = 0.05)
```

```python
model = MANOVA.from_formula('re_energy_con + CO2_em_pc + CO2_em_ppp + access_elec + access_clean_fuels ~ income_group', data = pollution_manova)
print(model.mv_test()) 
```
