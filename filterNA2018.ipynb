```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("wdi2018.xlsx")
```

```python
data
```

```python
data = data.replace(to_replace='..', value=np.NaN)
```

```python
data
```

```python
#% of missing values per column 

#Because division of sum by length is mean, you can instead df_missing.sum()/len(df) use df_missing.mean()
result = data.isna().mean()
```

```python
print(result)
```

```python
#remove columns with >05% NA
data_filtered95 = data.loc[:,result < .05]
#print(list(data_filtered95.columns))

#print(len(list(data_filtered95.columns)))
```

```python
#remove columns with >10% NA
data_filtered90 = data.loc[:,result < .1]
#print(list(data_filtered90.columns))

#print(len(list(data_filtered90.columns)))
```

```python
#remove columns with >20% NA
data_filtered80 = data.loc[:,result < .2]
#print(list(data_filtered80.columns))

print(len(list(data_filtered80.columns)))
```
